## Hospital Patient Managemnet By AVL(Trees) in Python DSA

```python
class Patient:
    def __init__(self, pid, name, date):
        self.id = pid
        self.name = name
        self.date = date

    def __str__(self):
        return f"ID:{self.id}, Name:{self.name}, Date:{self.date}"


class Node:
    def __init__(self, patient):
        self.patient = patient
        self.left = None
        self.right = None
        self.height = 1  


class AVLTree:
    def get_height(self, node):
        if not node:
            return 0
        return node.height

    def get_balance(self, node):
        if not node:
            return 0
        return self.get_height(node.left) - self.get_height(node.right)

    def right_rotate(self, y):
        x = y.left
        T2 = x.right

        x.right = y
        y.left = T2

        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))
        x.height = 1 + max(self.get_height(x.left), self.get_height(x.right))

        return x

    def left_rotate(self, x):
        y = x.right
        T2 = y.left

        y.left = x
        x.right = T2

        x.height = 1 + max(self.get_height(x.left), self.get_height(x.right))
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))

        return y

    def insert(self, root, patient):
        if not root:
            return Node(patient)
        elif patient.id < root.patient.id:
            root.left = self.insert(root.left, patient)
        else:
            root.right = self.insert(root.right, patient)

        root.height = 1 + max(self.get_height(root.left), self.get_height(root.right))

        balance = self.get_balance(root)

        if balance > 1 and patient.id < root.left.patient.id:
            return self.right_rotate(root)
        if balance < -1 and patient.id > root.right.patient.id:
            return self.left_rotate(root)
        if balance > 1 and patient.id > root.left.patient.id:
            root.left = self.left_rotate(root.left)
            return self.right_rotate(root)
        if balance < -1 and patient.id < root.right.patient.id:
            root.right = self.right_rotate(root.right)
            return self.left_rotate(root)

        return root

    def inorder(self, root):
        if root:
            self.inorder(root.left)
            print(root.patient)
            self.inorder(root.right)


# ---------------- DEMO ----------------
if __name__ == "__main__":
    tree = AVLTree()
    root = None

    root = tree.insert(root, Patient(3, "Ali", "2025-11-20"))
    root = tree.insert(root, Patient(1, "Sara", "2025-11-21"))
    root = tree.insert(root, Patient(2, "Ahmed", "2025-11-22"))
    root = tree.insert(root, Patient(4, "Zayn", "2025-11-23"))

    print("Patients in sorted order (by ID):")
    tree.inorder(root)
```
